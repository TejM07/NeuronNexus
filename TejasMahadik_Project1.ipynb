2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the
failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following
questions. [4 Marks]
  
  2.A. Probability that none of the items are defective? [1 Mark]
  
  2.B. Probability that exactly one of the items is defective? [1 Mark]
  
  2.C. Probability that two or fewer of the items are defective? [1 Mark]
  
  2.D. Probability that three or more of the items are defective ? [1 Mark]

```python

# Taking variable p as the failure rate for LED light bulbs that the company manufactures i,e, 5%
failure   =  0.05
# Taking n as the sample size of 10 bulbs
sample  =  10
# Arranging an array
array  =  np.arange(0,11)

#Using Probability mass function, we can get the probability of the bulbs being defective at each count of bulbs from 0 to 10
bi=stats.binom.pmf(array,sample,failure)

 #No defective items will be represented at the 0th position in the array
print('Solution to 2.A. : Probability that none of the items are defective is %1.5f' %bi[0])

print("")

 #Exactly one defective item will be represented at the 1st position in the array
print('Solution to 2.B. : Probability that exactly one of the items is defective is %1.5f' %bi[1])
print("")

#Using cumulative density function, we can get the cumulative probability of the bulbs being defective upto each count of bulbs from 0 to 10
cumbi=stats.binom.cdf(array,sample,failure)

print('Solution to 2.C. : Probability that two or fewer of the items are defective is %1.5f' %cumbi[2]) #this will calculate cumulative frequency of upto two bulbs being defective
print("")
print('Solution to 2.D. : Probability that three or more of the items are defective is %1.5f' %(1-cumbi[2])) #this will calculate cumulative frequency of three or more bulbs being defective by subtracting the cumulative frequency of upto two bulbs being defective from 1
```

3. A car salesman sells on an average 3 cars per week. [3 Marks]
  
  3.A. What is Probability that in a given week he will sell some cars? [1 Mark]
  
  3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]
  
  3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]

```python
avg = 3 #Average number of cars sold per week
#Calculate probabilities for each number of cars sold per week
n = np.arange(0,20)
poisson = stats.poisson.pmf(n,avg)

print("")

#The probability that he will sellsome cars means he shall sell atleast 1 car upto infinite number, hence to find the value we shall subtract the probability of selling no cars from 1
print('Solution to 3.A. : Probability that in a given week he will sell some cars is %1.5f' %(1-poisson[0]))
print("")

#The probability that he will sell 2 or more but less than 5 cars, we shall add the probabilities of selling 2,3 and 4cars
print('Solution to 3.B. : Probability that in a given week he will sell 2 or more but less than 5 cars is %1.5f' %np.sum(poisson[2:5]))
print("")

print('Solution to 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.')
plt.plot(n, poisson)

plt.title('Poisson Distribution: Probability Mass Function of Cars Sold per Week')
plt.xlabel('Number of Cars Sold per Week')
plt.ylabel('Probability of cars sold per-week')
plt.ylim(0.0000, 1.0000)  # Set y-axis limits from 0 to 1
plt.yticks(np.arange(0, 1, 0.05).round(decimals=4))
plt.grid(True)
plt.xticks(n)
plt.show()
```


4. Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed,
marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the
percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yours
independently place orders with the same bot. Answer the following questions. [3 Marks]

  4.A. What is the probability that all three orders will be recognised correctly? [1 Mark]

  4.B. What is the probability that none of the three orders will be recognised correctly? [1 Mark]

  4.C. What is the probability that at least two of the three orders will be recognised correctly? [1 Mark]


```python
#We will use the binomial probability concept for this. All events are independent of each other with success probability @ 86.8%

accuracyrate=0.868
notaccurate= 1-accuracyrate

eachcorrect= accuracyrate**3 #states each of the 3 orders is correct
nonecorrect = notaccurate**3  #states not a single order of the 3 orders was correct

print('Solution to 4.A. : Probability that all three orders will be recognised correctly is %1.5f' %eachcorrect)
print("")

print('Solution to 4.B. : Probability that none of the three orders will be recognised correctly is %1.5f' %nonecorrect)
print("")

# to find the 3rd answer we shall utilize the cumulative distribution function (cdf) of the binomial distribution and subtract the cumulative probability of getting 0 or 1 order correct from 1 to get the probability of at least two orders being correct.
# we can either use 1 - stats.binom.cdf(1,3,accuracyrate) OR stats.binom.pmf(2,3,accuracyrate) + stats.binom.pmf(3,3,accuracyrate)

a= 1 - stats.binom.cdf(1,3,accuracyrate)
print('Solution to 4.C. Probability that at least two of the three orders will be recognised correctly is %1.5f' %a)
```

B. Write the null hypothesis and alternative hypothesis

**Null Hypothesis (H0):**

There is no significant difference between the funds raised by companies that are still operating and companies that have closed down.

**Alternative Hypothesis (H1):**

There is a significant difference between the funds raised by companies that are still operating and companies that have closed down.


C. Test for significance and conclusion

```python
from statsmodels.stats.weightstats import ztest

sample1 = company1.loc[company1.OperatingState == 'Operating', 'Funds_in_million']
sample2 = company1.loc[company1.OperatingState =='Closed', 'Funds_in_million']

alpha = 0.05   # Let's consider a significance level of 5%

test_statistic, p_value = ztest(sample1, sample2)

if p_value <= alpha:
    print(f'Since the p-value i.e {round(p_value, 3)} < {alpha} i,e alpha the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value i.e {round(p_value,3)} > {alpha} i.e alpha the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:

**Null Hypothesis (H0):**
There is no significant difference in the proportion of companies that are operating between winners and contestants.

**Alternative Hypothesis (H1):**
There is a significant difference in the proportion of companies that are operating between winners and contestants.

H. Test for significance and conclusion

```python
# Perform a z-test for comparing two proportions
from statsmodels.stats.proportion import proportions_ztest

# Calculate the number of successes and total observations for winners and contestants
winners_success = company2[company2.Result == 'Winner'].OperatingState.value_counts().loc['Operating']
winners_total = company2.Result.value_counts()['Winner']
contestants_success = company2[company2.Result == 'Contestant'].OperatingState.value_counts().loc['Operating']
contestants_total = company2.Result.value_counts()['Contestant']

# Perform the z-test
z_statistic, p_value = proportions_ztest([winners_success, contestants_success], [winners_total, contestants_total])

# Set the significance level
alpha = 0.05

# Interpret the results
if p_value <= alpha:
    print('Reject the null hypothesis: There is a significant difference in the proportion of companies that are operating between winners and contestants.')
else:
    print('Fail to reject the null hypothesis: There is no significant difference in the proportion of companies that are operating between winners and contestants.')
```
